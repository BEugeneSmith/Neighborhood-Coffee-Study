```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
import re
```

```python
seattleBusinessData = pd.read_csv('data/seattle/Seattle_2015_Active_Business_License_Data.csv')
```

```python
seattleBusinessData.columns
```

```python
seattleBusinessData['Trade Name'] = seattleBusinessData['Trade Name'].str.lower()
seattleBusinessData['Business Legal Name'] = seattleBusinessData['Business Legal Name'].str.lower()
seattleBusinessData[['City','State','Zip']] = seattleBusinessData['City, State, ZIP'].str.split(',',expand=True)
seattleBusinessData[['State','Zip']] = seattleBusinessData['State'].str[1:].str.split(' ',expand=True)[[0,1]]
seattleBusinessData = seattleBusinessData[seattleBusinessData['City'] == 'SEATTLE']
```

```python
allCoffee = seattleBusinessData[seattleBusinessData['Trade Name'].str.contains('coffee')]
```
